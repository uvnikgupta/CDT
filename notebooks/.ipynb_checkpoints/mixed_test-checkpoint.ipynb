## This small section tests out the Mixed Discrete and Continuous type

```python
import pandas as pd
from scmodels import SCM
```

```python
mixed_scm = SCM(
    [
        # exogenous nodes
        "A1 = N, N ~ Bernoulli(p=0.8)",
        "A2 = N, N ~ Binomial(n=2, p=0.5, succ=2, fail=4)",
        "A3 = N, N ~ BetaBinomial(n=10, alpha=0.5, beta=2.2)",
        "A4 = N, N ~ Die(sides=6)",
        
        # first layer of dependency nodes
        "B1 = A1 + A2 + N,  N ~ DiscreteUniform(items='1 2 3')",
        "B2 = A1*A2 + A3*N, N ~ Geometric(p=0.8)",
        "B3 = A3 + A4*N,    N ~ Benini(alpha=0.1, beta=0.4, sigma=0.4)",
        "B4 = A3**2 - sqrt(A1 + A4*N), N ~ Poisson(lamda=0.8)",
        "B5 = N,  N ~ FiniteRV({10: 1/2, 0: 1/6, 23: 1/6, 31: 1/6})",
        
        # second layer of dependency nodes
        "C1 = A1*A3*0.5 + B1*N, N ~ Logarithmic(p=0.5)",
        "C2 = A1*A2 + B2*N, N ~ Beta(alpha=0.5, beta=0.5)",
        "C3 = A1*B3 + B3*N, N ~ Beta(alpha=0.5, beta=0.5)",
        "C4 = B1**2 + B5 + B4*N, N ~ Beta(alpha=0.5, beta=0.5)",
        
        # last layer of dependency nodes
        "D1 = A1 + A2 + C1*0.5*N, N ~ Normal(mean=2, std=1)",
        "D2 = B1 + B2 + C1*0.5*N, N ~ DiscreteUniform(items='1 0 2')",
        "D3 = C1 + C2 + C3*N,     N ~ Normal(mean=0, std=1)",
        "D4 = (A1 + B1 + C1) * N, N ~ Normal(mean=1, std=1)",
    ]
)
print(mixed_scm)
```

```python
mixed_scm.plot(node_size=1000, alpha=1)
```

```python
mixed_scm.sample(500)
```

### Do intervention at B5

```python
mixed_scm.do_intervention([("B5", 0)])
mixed_scm.sample(10)
```

```python
mixed_scm.undo_intervention()
```
